# 데이터 클리닝

axis=1 은 열제거

axis=0 은 행 제거

(아무것도 적지 않으면 axis=0이 디폴트)

```python
# Name 열 제거
# inplcae=True: 기존의 데이터 대체
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/titanic.csv')
data.drop('Name',axis=1,inplace=True)
data.head()
```

```python
# 행 지우기
data.drop(2, axis=0)
```

```python
# 두 개의 열을 동시에 drop할 때는 []으로 묶어줘야 함
data.drop(['Pclass','Age'],axis=1)
```

loc: 이름기반

iloc: 위치 기반

```python
# 특정 행 불러오기
data.loc[[4,5,6]]
```

```python
data.iloc[[4,5,6]]
```

```python
# 어떤 열에 결측치가 있는지 확인
data.isna().sum()
```

```python
# 특정 열 Age의 결측치 확인
data['Age'].isna()
```

```python
# True에 해당하는 값만 출력
age_na_index = data[data['Age'].isna()].index
age_na_index
```

```python
# 결측치 제거
data.dropna()
```

```python
# 특정 column의 결측치만 제거
data.dropna(subset = ['Embarked'])
```

mean(): 평균

median(): 중앙값

outlier(이상치)가 있을 경우에는 평균을 사용하면 평균이 지나치게 높아질 수 있음

```python
# mean으로 Age의 결측치를 채움
data['Age'].mean()
#data['Age'].fillna(29)
data['Age'] = data['Age'].fillna(data['Age'].mean())
data.loc[age_na_index]
```

object type의 value의 결측치를 채워줄 때

1) 가장 많은 값으로 채움

2) Null 값을 명시해줌

```python
# Embarked 결측치 채움
# 각각 값이 몇 개인지 알 수 있음
data['Embarked'].value_counts()
data['Embarked'].fillna('S')
```

```python
# Ticket 값들을 띄어쓰기 기준으로 분할 후 마지막 것들만 가져옴
data['new_ticket'] = data['Ticket'].str.split().apply(lambda x : x[-1])
data
```

```python
# LINE을 999999로 대체
data['new_ticket'] = data['new_ticket'].replace({'LINE':'999999'})
```

```python
# data 타입을 'int'로 바꿔줌
data['new_ticket'].astype('int')
```

```python
# to_numeric을 사용해 int로 변환
data['new_ticket'] = pd.to_numeric(data['new_ticket'])
```

```python
# new_ticket 컬럼이 int로 바뀐 것을 확인
data.info()
```

```python
from datetime import datetime
```

```python
# strptime: 문자열(String)을 날짜(Date) 객체로 바꿔주는 함수
sample_date = datetime.strptime('2023-01-01','%Y-%m-%d')
sample_date
```

```python
datetime.strftime(sample_date,'%Y-%m-%d')
```

```python
sample_date.year
```

```python
# 두 날짜 간의 차이
next_date = datetime.strptime('2023-05-05','%Y-%m-%d')
next_date - sample_date
```

```python
# 형태 datetime으로 바꾸기
data['ticket_date'] = pd.to_datetime(data['ticket_date'])
```

```python
data.info()
```

```python
# year만 불러올 때
data['ticket_date'].dt.year
```

```python
# 사고 날짜와 티켓 구매 날짜의 차이
acc = datetime.strptime('1912-04-15', '%Y-%m-%d')
```
