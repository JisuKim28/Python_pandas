```python
# 열 n개 선택 = 데이터프레임 객체 반환
# 2중 대괄호([[]]) 안에 열 이름을 따옴표와 함께 입력
# 'sex', 'age' column만 추출하여 재구성
passenger = titanic[['Sex','Age']]
passenger.head()
```

```python
print(type(passenger))
print(passenger.shape)
```

```python
# 조건: 35살 초과인 데이터 추출(True,False 반환)
print(passenger['Age'] > 35)
```

```python
# 조건: 35살 초과인 데이터 추출
# 불리언 인덱싱: 조건의 결과가 True인 값만 추출
above35 = passenger[passenger['Age']>35]
above35.head()
```

```python
# 데이터 내용 확인
titanic.head(2)
```

```python
# .isin 함수는 각각의 요소가 데이터프레임 또는 시리즈에 존재하는지 파악해 T/F 반환
# Pclass 변수의 값이 1일 경우, T/F 반환
titanic['Pclass'].isin([1])
```

```python
# 불리언 인덱싱 + isin(): 데이터 특정 범위만 추출
# Pclass 변수의 값이 1일 경우만 추출해 class1에 저장
class1 = titanic[titanic["Pclass"].isin([1])]
class1.head()
```

```python
passenger.head(2)
```

```python
# 불리언 인덱싱 + .isin(): 데이터의 특정 범위만 출력
# Age 변수의 값이 20살부터 40살까지의 범위만 추출하여 age2040에 저장
age2040 = passenger[passenger['Age'].isin(np.arange(20, 41))]
age2040.head()
```

```python
# 결측값 파악을 위한 데이터 확인
passenger.head(7)
```

```python
# .isna(): 결측값은 True 반환, 그 외에는 False 반환
# .isna 함수는 데이터프레임 내에 결측 값을 확인하기 위해 사용
passenger['Age'].isna()[0:7]
```

```python
# 결측값 확인
ages = passenger[passenger['Age'].isna()]
ages.head(2)
```

```python
# 결측값 파악을 위한 데이터 확인
passenger.head(7)
```

```python
# .notna(): 결측값은 False 반환, 그 이외에는 True
# .notna 함수는 누락되지 않은 값을 찾기 위해 사용
passenger['Age'].notna()[0:7]
```

```python
# 결측값을 제거한 누락되지 않은 값을 확인
ages = passenger[passenger["Age"].notna()]
ages.head(7)
```

```python
# 결측값 파악을 위한 데이터 확인
titanic.head(3)
```

```python
# .dropna(axis=0): 결측값이 들어있는 행 전체 삭제
titanic.dropna(axis=0).head(3)
```

```python
# .dropna(axis=1): 결측값이 들어있는 열 전체 삭제
titanic.dropna(axis=1).head(3)
```

```python
# .loc[]: 행 이름과 열 이름을 사용
# 나이가 35세 초과인 사람의 이름과 나이 출력
names35 = titanic.loc[titanic['Age'] > 35, ['Name','Age']]
names35.head()
```

```python
# .iloc[]: 행 번호와 열 번호를 사용
# 1번째 행부터 3번째행까지의 0번째 열의 값을 "No Name" 변경
names35.iloc[[1,2,3],0] = 'No Name'
names35.head()
```

```python
# Age 변수의 평균 구하기
print(titanic['Age'].mean())
```

```python
# Age 변수의 중앙값 구하기
print(titanic['Age'].median())
```

```python
# descrbe() 함수를 통해 다양한 통계량 요약
# 특정 변수 'Age', 'Fare' 추출
print(titanic[['Age','Fare']].describe())
```

```python
# agg()를 통해 여러 개의 열에 다양한 함수를 적용
# 모든 열에 여러 함수를 매핑
# 각 열마다 다른 함수를 매핑
print(titanic.agg({'Age':["min","max","median","std"],
                   'Fare': ["min",'max','mean','median']}))
print("그다음")
print(titanic.agg({'Age':'mean','Fare':'median'}))
```

```python
# groupby() 함수를 이용하여 그룹별 집계 <예제1>
# grouupby로 성별과 클래스를 묶어주고 나이와 요금의 평균 구하기
titanic.groupby(['Sex','Pclass'])[['Age','Fare']].mean()
```

```python
# groupby() 함수를 이용하여 그룹별 집계 <예제2>
# groupby를 통해 성별을 묶은 다음, 생존율의 평균 구하기
survive = titanic.groupby('Sex')['Survived'].mean()
(survive * 100).head()
```

```python
# value_counts() 함수를 이용하여 개수 ㄱ하기
titanic['Pclass'].value_counts()
```
