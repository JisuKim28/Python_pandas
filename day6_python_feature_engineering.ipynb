```python
# 컬럼 이름 변경
# pref_o: 상대방이 각각의 요소에 대해 얼마나 중요하게 생각하는가 (합: 100)
# _o: 상대방의 평가
# _important: 나에게 각 요소를 얼마나 중요하게 생각하는가
# _parner: 내가 상대방을 평가
dating_df.columns
```

```python
# 내가 평가한 항목은 i, 상대방이 평가한 항목은 o로 바꿔줌
'string'.startswith('s')
'string'.endswith('j')
```

```python
new_cols = []
for i in dating_df.columns:
  if i.startswith('pref_o'):
    i = i.replace('pref_o','o_important')
  elif i.endswith('_o'):
    i = 'o_score_' + i.replace('_o','')
  elif i.endswith ('important'):
    i = 'i_important_' + i.replace('_important','')
  elif i.endswith('_partner'):
    i = 'i_score_' + i.replace('_partner','')
  new_cols.append(i)
```

```python
dating_df.columns = new_cols
```

사람마다 다르게 준 중요도 점수를 "같은 기준"으로 바꾸는 작업

사람마다 준 점수 총합이 다르기 때문에, 같은 기준(합 100) 으로 맞춰서 공정하게 비교하려고

(100 / 총합) * 각 점수 식으로 정규화

```python
# 모든 사람들이 합이 100이 되는 조건을 만들어줌 -> 가중치를 통해
# (100/sum) * 95 <-- 각 항목의 값
# sum이 100일 경우에는 가중치가 없기 때문에 결과가 1이 나옴
dating_df[o_imp] = dating_df.apply(lambda x : (100 / x['o_imp_sum']) * x[o_imp] , axis=1)
```

```python
dating_df[o_imp].sum(axis=1)
```

```python
dating_df[i_imp] = dating_df.apply(lambda x : (100 / x['i_imp_sum']) * x[i_imp] , axis=1)
```

```python
dating_df[i_imp].sum(axis=1)
```

```python
# o_imp_sum, i_imp_sum 두 컬럼의 합을 100으로 맞춰뒀기 때문에 더이상 필요 없음 -> 삭제
dating_df.drop(['o_imp_sum','i_imp_sum'], axis=1, inplace=True)
```

# Age

```python
def age_func(x):
  if x['age'] == -99:
    return -99
  elif x['age_o'] == -99:
    return -99
  elif x['gender'] == 'female':
    return x['age_o'] - x['age']
  else:
    return x['age'] - x['age_o']
```

```python
dating_df['age_gap'] = dating_df.apply(age_func, axis=1)
```

```python
dating_df['age_gap']
```

```python
# 남자와 여자 중 누가 더 나이가 많은지 판단하기 위함
dating_df['age_gap_dir'] = dating_df['age_gap'].apply(lambda x : 'positive' if x > 10 else 'negative' if x < 0 else 'zero')
```

```python
# 나이차 절댓값 처리
dating_df['age_gap'] = abs(dating_df['age_gap'])
```

# Rating

```python
# i_important, i_score, o_imporatnt, o_score 각각의 컬럼을 리스트로 모음
o_important = []
o_score = []
i_important = []
i_score = []

for i in dating_df.columns:
  if i.startswith('o_important'):
    o_important.append(i)
  elif i.startswith('o_score'):
    o_score.append(i)
  elif i.startswith('i_important'):
    i_important.append(i)
  elif i.startswith('i_score'):
    i_score.append(i)
```

```python
# o_important의 0값을 99로 변경
dating_df[o_important].replace({0:-99})
```

```python
# o_important의 0값을 99로 변경
dating_df[i_important].replace({0:-99})
```

```python
# o_important_attractive * o_score_attractive
def rating(data, important, score):
  if data[score] == -99:
    return -99
  elif data[important] == -99:
    return -99
  else:
    return data[important] * data[score]
```

```python
dating_df.apply(lambda x : rating(x, 'o_important_attractive','o_score_attractive'), axis=1)
```
